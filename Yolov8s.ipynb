```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt")  # nceden eitilmi model

model.train(
    data="/content/data.yaml",
    epochs=100,
    imgsz=960,
    batch=16,
    workers=2,
    device=0,        # GPU varsa 0, yoksa CPU zaten otomatik
    pretrained=True, # transfer learning
    #augment=True,    # YOLO yerleik augmentation
    #mosaic=0.3,     # 
    #mixup=0.05,     # 
    #translate=0.1,
    #scale=0.15,
    #fliplr=0.3,
    patience=20,      # erken durdurma
    lr0=0.001,
    cos_lr=True

)
```

```python
from ultralytics import YOLO

model = YOLO("/content/runs/detect/train/weights/best.pt")

results = model.val(data="/content/data.yaml", split="test")

metrics = results.box

# Genel metrikler
print("General Metrics")
print("Precision:", metrics.p.mean())
print("Recall:", metrics.r.mean())
print("mAP50:", metrics.map50)
print("mAP50-95:", metrics.map)
```
